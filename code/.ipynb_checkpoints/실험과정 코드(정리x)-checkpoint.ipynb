# knn 실험

```python
x_train, x_test, y_train, y_test, x, y = prep('E:/workplace/AMD/drebin-실험1.csv',0.2,7)
from sklearn.neighbors import KNeighborsClassifier
# N-fold cross-validation
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

kclf = KNeighborsClassifier(n_neighbors=5)
kclf.fit(x_train,y_train)

print('knn score :',kclf.score(x_test,y_test),'\n')
mean = 0
for i in cross_val_score(kclf, x,y, scoring=None, cv=10):
    mean += i
print('knn 10 fold score :',mean/10,'\n')

x_tr, x_te, y_tr, y_te, x_, y_ = prep('E:/workplace/AMD/amd-실험1.csv',0.2,7)
print(kclf.score(x_, y_))

# confusion_matrix(y,kclf.predict(x))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100,max_depth=20)
clf.fit(x_train,y_train)

print('rf score :',clf.score(x_test,y_test),'\n')
mean = 0
for i in cross_val_score(clf, x,y, scoring=None, cv=10):
    mean += i
print('rf 10 fold score :',mean/10,'\n')

# x_tr, x_te, y_tr, y_te, x_, y_ = prep('E:/workplace/AMD/drebin-실험1.csv',0.2,7)
print(clf.score(x_, y_))

# confusion_matrix(y,clf.predict(x))
```

```python
def amdex(amdset):
    x_train, x_test, y_train, y_test, x, y = prep('E:/workplace/AMD/'+amdset+'.csv',0.2,7)
    del x_train['Unnamed: 0']
    del x_test['Unnamed: 0']
#     del y_train['Unnamed: 0']
#     del y_test['Unnamed: 0']
    del x['Unnamed: 0']
#     del y['Unnamed: 0']
    
    kclf = KNeighborsClassifier(n_neighbors=5)
    kclf.fit(x_train,y_train)

    print('knn score :',kclf.score(x_test,y_test),'\n')

    mean = 0
    for i in cross_val_score(kclf, x,y, scoring=None, cv=10):
        mean += i
    print('knn 10 fold score :',mean/10,'\n')


    clf = RandomForestClassifier(n_estimators=100,max_depth=20)
    clf.fit(x_train,y_train)

    print('rf score :',clf.score(x_test,y_test),'\n')
    mean = 0
    for i in cross_val_score(clf, x,y, scoring=None, cv=10):
        mean += i
    print('rf 10 fold score :',mean/10,'\n')

    joblib.dump(kclf,amdset+'_knn')
    joblib.dump(clf,amdset+'_rf')
```

```python
import lightgbm as lgb
from sklearn.model_selection import StratifiedKFold
from prep import *
from sklearn.metrics import mean_squared_error,roc_auc_score,precision_score

x_train, x_test, y_train, y_test, x, y = prep('E:/workplace/AMD/amd-실험1.csv',0.2,7)

x_train2, x_test2, y_train2, y_test2, x2, y2 = prep('E:/workplace/AMD/drebin-실험1.csv',0.2,7)

para = {'max_depth':5,
        'objective': 'binary',
        'boosting_type' :'gbdt',
        'learning_rates':0.1,
        'num_leaves':2 ** 7 - 1,
        'bagging_fraction':0.7,
        'feature_fraction':0.7,
        'verbose_eval':1,
        'n_jobs':8,
        'seed':4
        }

pred = [0] * len(x2)
fold_num = 5
valid_accs=[]
test_accs = []
folds = StratifiedKFold(n_splits=fold_num, shuffle=True, random_state=4)

for fold_, (train, valid) in enumerate(folds.split(x, y)):
    print(f"{fold_+1} fold")
    x_train, y_train = x.iloc[train][x.columns], y.iloc[train]
    x_test, y_test = x.iloc[valid][x.columns], y.iloc[valid]
    
    train_ds = lgb.Dataset(x_train, label=y_train)
    val_ds = lgb.Dataset(x_test, label=y_test)
    
    lgb_train = lgb.train(para, train_ds,  100, val_ds, verbose_eval=10)
    
    y_pred = lgb_train.predict(x_test)
    y_pred = y_pred.round(0)
    y_pred = y_pred.astype(int)
    valid_accs.append(roc_auc_score(y_pred,y_test))
    
    y_pred2 = lgb_train.predict(x2)
    y_pred2 = y_pred2.round(0)
    y_pred2 = y_pred2.astype(int)
    test_accs.append(roc_auc_score(y_pred2,y2))

print('\n 5차 교차검증 validation set 평균 정확도 : ', str(sum(valid_accs)/fold_num))
print('\n 5차 교차검증 test set 평균 정확도 : ', str(sum(test_accs)/fold_num))
```
